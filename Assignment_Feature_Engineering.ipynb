# **Assignment Feature Engineering**

Q1. What is a parameter ?

Ans. A parameter is a model’s internal variable learned during training.
It helps the model make predictions (e.g., weights in linear regression).
Parameters are optimized to minimize the loss function.
They differ from hyperparameters, which are set manually.

Q2. What is correlation?

Ans.Correlation measures the relationship between two variables.
It ranges from -1 (negative) to +1 (positive).
A value near 0 means no linear correlation.
It's useful in feature selection during EDA.

Q3. What does negative correlation mean?

Ans. As one variable increases, the other decreases.
Represented by values between 0 and -1.
Example: More study hours → Less entertainment time.
It's important in identifying inverse relationships.

Q4. Define Machine Learning. What are the main components in Machine Learning?

Ans. ML enables systems to learn from data without being explicitly programmed.
Components:

Data – Input to train the model.

Model – Learns from data and makes predictions.

Loss Function – Measures error in prediction.

Optimizer – Updates parameters to minimize loss.

Q5. How does loss value help in determining whether the model is good or not?

Ans. Loss indicates the error between predicted and actual values.
Lower loss means the model is performing better.
High loss suggests underfitting or wrong model choice.
It helps in comparing different models or tuning.

Q6. What are continuous and categorical variables?

Ans.Continuous: Numeric with infinite values (e.g., height, age).
Categorical: Fixed categories/labels (e.g., gender, city).
Categorical can be ordinal (ordered) or nominal (unordered).
Understanding variable type is crucial for preprocessing.

Q7. How do we handle categorical variables in Machine Learning? What are the common techniques?

Ans. Categorical data must be encoded into numeric values.
Common methods:
Label Encoding – Assigns a unique integer to each category.
One-Hot Encoding – Creates binary columns for each category.
Scikit-learn offers tools like LabelEncoder, OneHotEncoder.

Q8. What do you mean by training and testing a dataset?

Ans. Training set is used to train the model and learn patterns.
Testing set is used to evaluate model performance on unseen data.
Helps prevent overfitting and assess generalization.
Usually split as 70% training and 30% testing (or similar).

Q9. What is sklearn.preprocessing?

Ans.It’s a Scikit-learn module for preparing raw data for modeling.
Includes functions for scaling, encoding, normalizing, etc.
Examples: StandardScaler, MinMaxScaler, LabelEncoder.
Ensures that input data is properly formatted for ML models.

Q10. What is a Test set?

Ans. It's the part of the dataset used only for evaluation.
It simulates how the model will perform on real-world data.
Not involved in training at all.
Accuracy, precision, recall, and loss are often calculated on it.

Q11. How do we split data for model fitting (training and testing) in Python?

Ans. We use train_test_split() from sklearn.model_selection.
Typical test sizes: 20% or 30% of the dataset.
Ensures a fair evaluation of model performance.
Example:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

Q17. What is causation? Explain difference between correlation and causation with an example.

Ans. Causation means one variable directly affects another.
Correlation means variables move together but may not be linked.
Example:
Correlation: Ice cream sales ↑, drowning ↑ (both due to summer).
Causation: Exercise ↑ → Weight ↓.

Q18. What is an Optimizer? What are different types of optimizers?

Ans. Optimizers adjust model parameters to minimize loss.
SGD (Stochastic Gradient Descent): Updates with one/few samples.
Adam: Adaptive learning rates; faster convergence.
RMSprop: Good for RNNs, divides learning rate by moving average of gradients.

Q19. What is sklearn.linear_model?

Ans. A module in Scikit-learn for linear models.
Includes: LinearRegression, LogisticRegression, Ridge, etc.
These models assume a linear relationship between features and target.
Easy to implement and interpret.

Q23. What is feature scaling? How does it help in Machine Learning?

Ans. Standardizes the range of features.
Helps models like SVM, KNN, and Gradient Descent converge faster.
Prevents features with large values from dominating.
Common methods: Min-Max Scaling, Standardization.

Q24. How do we perform scaling in Python?

Ans. Use StandardScaler or MinMaxScaler from Scikit-learn.
StandardScaler: Mean = 0, Std = 1
MinMaxScaler: Scales values between 0 and 1
Example:
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
X_scaled = scaler.fit_transform(X)
